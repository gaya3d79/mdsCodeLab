### **4. Standardise Datasets**

**4.1 Institutional Level Data**

The IPEDS enrollment files contain a column referring to the 'Level of student' as well as aggregate totals. For the purpose of this study the column is renamed to 'Level_of_student' and only records representing 'All students total' are retained as the analysis focuses on institution-level totals.

```python
def fix_student_col(df):
    for col in df.columns:
        if "Level of student" in col:
            df = df.rename(columns={col: "Level_of_student"})
    # Keep only 'All students total'
    df = df[df['Level_of_student'] == 'All students total']
    return df
```

**4.2 Student Level Data**

Keep only the rows where 'Level of Student' is 'All students total' based on the standardized column name related to 'Level of student'. This is to ensure consistency.

```python
Gender2019 = fix_student_col(Gender2019)
Race2019 = fix_student_col(Race2019)

Gender2020 = fix_student_col(Gender2020)
Race2020 = fix_student_col(Race2020)

Gender2021 = fix_student_col(Gender2021)
Race2021 = fix_student_col(Race2021)

Gender2022 = fix_student_col(Gender2022)
Race2022 = fix_student_col(Race2022)

Gender2023 = fix_student_col(Gender2023)
Race2023 = fix_student_col(Race2023)
```

**5. Yearly IPEDS Data Integration**

The IPEDS files are merged for each year from 2019 to 2023 using common UNITID, institution name and year fields. This combines enrollment, demographic, graduation, admissions and administrative variables into one dataset for each year. An outer join is used to preserve available records from all source files, reducing the risk of losing institutions due to partial data availability.

```python
# merge 2019
merged2019 = Gender2019.merge(
    Race2019,
    on=["unitid","institution name", "year", "Level_of_student"],
    how="outer"
).merge(
    GraduationRates2019,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    SingleFileAdm2019,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    admin2019,
    on=["unitid","institution name", "year"],
    how="outer"
)

# merge 2020 files
merged2020 = Gender2020.merge(
    Race2020,
    on=["unitid","institution name", "year", "Level_of_student"],
    how="outer"
).merge(
    GraduationRates2020,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    SingleFileAdm2020,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    admin2020,
    on=["unitid","institution name", "year"],
    how="outer"
)

# merge 2021
merged2021 = Gender2021.merge(
    Race2021,
    on=["unitid","institution name", "year", "Level_of_student"],
    how="outer"
).merge(
    GraduationRates2021,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    SingleFileAdm2021,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    admin2021,
    on=["unitid","institution name", "year"],
    how="outer"
)

# merge 2022
merged2022 = Gender2022.merge(
    Race2022,
    on=["unitid","institution name", "year", "Level_of_student"],
    how="outer"
).merge(
    GraduationRates2022,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    SingleFileAdm2022,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    admin2022,
    on=["unitid","institution name", "year"],
    how="outer"
)

# merge 2023
merged2023 = Gender2023.merge(
    Race2023,
    on=["unitid","institution name", "year", "Level_of_student"],
    how="outer"
).merge(
    GraduationRates2023,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    SingleFileAdm2023,
    on=["unitid","institution name", "year"],
    how="outer"
).merge(
    admin2023,
    on=["unitid","institution name", "year"],
    how="outer"
)
```

**6. Column Standardisation and Restructuring**

Column names are standarised for consistency across years in IPEDS datasets. This removes the year specific suffixes, formating diferrences, long labels and then map similar variables into unified column names.

```python
import re

def standardize_columns(df):

    new_cols = []

    for col in df.columns:
        c = col.lower().strip()

        #remove years in the column title (2019, 2020, etc.)
        c = re.sub(r'\d{2,4}', '', c)

        #remove suffixes
        c = c.replace('_rv', '')
        c = c.replace('_x', '')
        c = c.replace('_y', '')

        #replace dots & spaces
        c = c.replace('.', '_')
        c = c.replace(' ', '_')

        #clean underscores
        c = re.sub(r'_+', '_', c)

        #rename columns

        #enrollment (EF)
        if 'ef_total_men' in c:
            c = 'enrollment_men'
        elif 'ef_total_women' in c:
            c = 'enrollment_women'

        #race (EFA)
        elif 'asian' in c:
            c = 'enrollment_asian'
        elif 'black' in c:
            c = 'enrollment_black'
        elif 'hispanic' in c:
            c = 'enrollment_hispanic'
        elif 'white' in c:
            c = 'enrollment_white'
        elif 'nonresident' in c:
            c = 'enrollment_international'

        #graduation (GR)
        elif 'grand_total' in c and 'gr' in col.lower():
            c = 'graduation_total'
        elif 'gr_total_men' in c:
            c = 'graduation_men'
        elif 'gr_total_women' in c:
            c = 'graduation_women'

        #admissions (ADM)
        elif 'applicants_total' in c:
            c = 'applicants_total'
        elif 'applicants_men' in c:
            c = 'applicants_men'
        elif 'applicants_women' in c:
            c = 'applicants_women'

        elif 'admissions_total' in c:
            c = 'admissions_total'
        elif 'admissions_men' in c:
            c = 'admissions_men'
        elif 'admissions_women' in c:
            c = 'admissions_women'

        elif 'enrolled_total' in c:
            c = 'enrolled_total'
        elif 'enrolled_men' in c:
            c = 'enrolled_men'
        elif 'enrolled_women' in c:
            c = 'enrolled_women'

        #SAT/ACT
        elif 'sat_math_25' in c:
            c = 'sat_math_25'
        elif 'sat_math_75' in c:
            c = 'sat_math_75'
        elif 'sat_evidence' in c and '25' in c:
            c = 'sat_reading_25'
        elif 'sat_evidence' in c and '75' in c:
            c = 'sat_reading_75'
        elif 'act_composite_25' in c:
            c = 'act_25'
        elif 'act_composite_75' in c:
            c = 'act_75'

        #financial aid
        elif 'financial' in c or 'aid' in c:
            if 'any' in c:
                c = 'aid_any_percent'
            elif 'pell' in c:
                c = 'aid_pell_percent'
            elif 'average' in c:
                c = 'aid_avg_amount'

        #retention/faculty
        elif 'retention' in c:
            c = 'retention_rate'
        elif 'faculty' in c:
            c = 'student_faculty_ratio'

        #enrollment breakdown
        elif 'full_time_enrollment' in c:
            c = 'enrollment_full_time'
        elif 'undergraduate_enrollment' in c:
            c = 'enrollment_undergrad'
        elif 'graduate_enrollment' in c:
            c = 'enrollment_grad'

        #institution info
        elif 'control_of_institution' in c:
            c = 'institution_control'
        elif 'urbanization' in c:
            c = 'urbanization'
        elif 'carnegie' in c:
            c = 'carnegie_classification'

        new_cols.append(c)

    df.columns = new_cols
    return df
```

```python
#column standisation across years
df2019 = standardize_columns(df2019)
df2020 = standardize_columns(df2020)
df2021 = standardize_columns(df2021)
df2022 = standardize_columns(df2022)
df2023 = standardize_columns(df2023)
```

```python
#remove duplicate columns if different labels are mapped into the same variable name
def remove_duplicate_columns(df):
    return df.loc[:, ~df.columns.duplicated()]

df2019 = remove_duplicate_columns(df2019)
df2020 = remove_duplicate_columns(df2020)
df2021 = remove_duplicate_columns(df2021)
df2022 = remove_duplicate_columns(df2022)
df2023 = remove_duplicate_columns(df2023)
```

**7. IPEDS Multiple Years Data Integration**

Integrated the 2019-2023 datasets into a single dataset

```python
combined = pd.concat([df2019, df2020, df2021, df2022, df2023], ignore_index=True)
```

```python
pivot_df = combined.pivot_table(
    index=["unitid", "institution_name", "year"],
    columns="gr_cohort_data",
    values=[
        "graduation_total",
        "graduation_men",
        "graduation_women"
    ],
    aggfunc="first"
).reset_index()
```

```python
pivot_df.columns = [
    "_".join(col).strip("_") if isinstance(col, tuple) else col
    for col in pivot_df.columns
]
```

```python
base = combined.drop_duplicates(
    subset=["unitid", "institution_name", "year"]
)
```

```python
base = base.drop(columns=[
    "gr_cohort_data",
    "graduation_total",
    "graduation_men",
    "graduation_women"
])
```

```python
final_df = base.merge(
    pivot_df,
    on=["unitid", "institution_name", "year"],
    how="left"
)
```

```python
final_df = final_df.sort_values(
    by=["unitid", "institution_name", "year"]
)
```
